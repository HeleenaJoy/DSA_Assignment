# Check whether all the zones generate the same amount of sales.


```python
#Set Hypothesis
#H0:Null Hypothesis:all the zones generate the same amount of sales.
#H1:Alternate Hypothesis:all the zones do not generate the same amount of sales.
```

```python
f,p = stats.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
round(f,5)
```

```python
round(p,5)
```

Since p<0.05 , we are Rejecting Null Hypothesis. that means,all the zones do not generate the same amount of sales.
