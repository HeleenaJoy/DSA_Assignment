# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
#Plotting count of male and female buyers who participated in sales.
plt.figure(figsize=(7,5))
cars['Buyer Gender'].value_counts().plot(kind='bar')
plt.title('Male Vs Female buyers participated in the sales',fontsize=14)
plt.xlabel('Buyer Gender',fontsize=12)
plt.ylabel('Number of participants',fontsize=12)
```

From graph it is clear that Female count is slightly greater than male count.

# 4. Find the top 5 cars based on their sales price.

```python
#sorting data in descending order
car=cars.sort_values('Sale Price',ascending=False)
car.head()
```

```python
car[['Make','Model']].head()
```

Above shown are the top 5 cars. That is Volkswagen>Audi>Lexus -Model LX>Mazda>Lexus-Model ES
